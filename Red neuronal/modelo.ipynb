```python

# Seleccionamos columnas numéricas para clustering
pop_features = pop2_df.select_dtypes(include=['int64', 'float64'])

# Aplicamos KMeans con 10 clusters
kmeans = KMeans(n_clusters=10, random_state=42)
pop2_df['genre_cluster'] = kmeans.fit_predict(pop_features)
```

```python
# Limpiar columnas
pop_df.columns = pop_df.columns.str.strip()

# Crear mapa de artista -> cluster
artist_genre_map = pop_df.copy()
artist_genre_map['genre_cluster'] = kmeans.predict(
    pop_df.select_dtypes(include=['int64', 'float64'])
)
artist_genre_map = artist_genre_map.reset_index(drop=True)

# Renombrar para igualar nombres
artist_genre_map = artist_genre_map.rename(columns={'artists': 'artist'})

train_df_labeled = train_df.merge(
    artist_genre_map[['artist', 'genre_cluster']],
    on='artist',
    how='inner'
)

```
